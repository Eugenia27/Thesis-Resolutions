```python
def una_componente(r, i_s, r_s, beta):
    return i_s*np.exp(-(r/r_s)**beta)
```

```python
def sersic_caon_ie(r,ie,re,beta):
    b= -0.142+0.868*1./beta 
    return ie*np.exp(-b*((r/re)**beta -1))
```

```python
def mu(_ie, _delta_band):
    return (-2.5*np.log10(_ie/4.2545e10)-48.6 + _delta_band )
```

```python
#Modelo de Sersic ----> Charly supermongo
def mufit(_r, _mu0, _rs, _beta):
    cte = 1.0857
    return _mu0 + cte*(_r/_rs)**_beta
```

```python
#Modelo de Sersic ----> N. Caon+ 93. On the shape of the light profiles of early-type galaxies. Con el beta del charly
def mufit2(_r, _mue, _re, _beta):
    return _mue + 2.5*(0.868/_beta-0.142)*((_r/_re)**_beta)
    #return _mue + (2*(1./_beta)-0.33)*((_r/_re)**_beta)
```
